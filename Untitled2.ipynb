```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("notas_estudiantes.csv")

df.head()

promedio=df.groupby('asignatura')["nota"].mean()
df.head()

plt.figure(figsize=(8, 5))
plt.bar(promedio.index, promedio.values, color="purple", edgecolor="green")


plt.title("Nota Promedio por Materia")
plt.xlabel("Materia")
plt.ylabel("Media de los estudiante")

plt.ylim(0, 8)  # Ajusta según el rango de notas
plt.grid(axis='y', linestyle='-', alpha=1)
plt.tight_layout()

# Mostrar
plt.show()

#estudiante
estudiante1 = 'Estudiante_2'
df_est = df[df['estudiante'] == estudiante1]

plt.figure(figsize=(8, 5))
plt.plot(df_est['asignatura'], df_est['nota'], marker='o', linestyle='-', color='orange')
plt.title(f"Notas de {estudiante1}")
plt.xlabel("Asignatura")
plt.ylabel("Nota")
plt.ylim(0, 8)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
